```python
from confluent_kafka import Producer

# Configuration for Confluent Cloud
conf = {
    'bootstrap.servers': 'pkcxxxx9092',  # from Confluent Cloud cluster settings
    'security.protocol': 'SASL_SSL',
    'sasl.mechanism': 'PLAIN',
    'sasl.username': 'xxx',     # from Confluent Cloud
    'sasl.password': 'xxx',  # from Confluent Cloud
}

# Create Producer instance
producer = Producer(conf)

# Delivery report callback
def delivery_report(err, msg):
    if err is not None:
        print(f"Delivery failed: {err}")
    else:
        print(f"Message delivered to {msg.topic()} [{msg.partition()}] at offset {msg.offset()}")

# Send messages
for i in range(5):
    producer.produce(
        'sctp_3',
        key=f'key-{i}',
        value=f'hello world hello sctp {i}',
        callback=delivery_report
    )

# Flush to ensure delivery
producer.flush()
```
