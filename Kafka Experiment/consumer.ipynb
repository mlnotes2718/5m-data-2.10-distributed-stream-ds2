```python
import streamlit as st
from confluent_kafka import Consumer, KafkaError

# Kafka Consumer configuration
conf = {
    'bootstrap.servers': 'pkcxxxx9092',   # from Confluent Cloud
    'security.protocol': 'SASL_SSL',
    'sasl.mechanism': 'PLAIN',
    'sasl.username': 'xxxx',     # from Confluent Cloud
    'sasl.password': 'xxx',  # from Confluent Cloud
    'group.id': 'streamlit-consumer',           # unique group id
    'auto.offset.reset': 'earliest'             # start from earliest if no committed offset
}

consumer = Consumer(conf)
consumer.subscribe(['topic_0_sctp'])

st.title("📡 Kafka Consumer Dashboard")

# Button to poll messages
if st.button("Poll Kafka Messages"):
    messages = []
    for _ in range(5):   # poll 5 messages per click
        msg = consumer.poll(1.0)
        if msg is None:
            continue
        if msg.error():
            if msg.error().code() != KafkaError._PARTITION_EOF:
                st.error(f"Kafka error: {msg.error()}")
        else:
            messages.append(msg.value().decode('utf-8'))

    if messages:
        st.write("### New Messages:")
        for m in messages:
            st.success(m)
    else:
        st.warning("No new messages found.")

consumer.close()
```
